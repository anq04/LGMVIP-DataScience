```python
x = iris_dset.drop(['class'], axis=1)
y = iris_dset['class']
print(f'x shape: {x.shape} | y shape: {y.shape} ')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
model= []
model.append(LogisticRegression(max_iter=3000))
model.append(LinearDiscriminantAnalysis())
model.append(KNeighborsClassifier())
model.append(DecisionTreeClassifier())
model.append(GaussianNB())

for modl in model:
    m_fit=modl.fit(x_train,y_train)
    predict_y=m_fit.predict(x_test)
    accuracy=accuracy_score(y_test,predict_y)
    matrix=confusion_matrix(y_test,predict_y)
    report=classification_report(y_test,predict_y)
    print(str(m_fit)+'\nAccuracy:'+str(accuracy)+'\nConfusion Matrix:'+str(matrix)+'\nClassification Report:'+str(report))
    
```

Conclusion: Linear Discriminant Analysis has the highest accuracy
