```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r'C:\Users\Computers way\Downloads\photo.jpg')
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
```

```python
plt.imshow(image)
plt.axis(False)
plt.show()
```

```python
RGBimg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGBimg)
plt.axis(False)
plt.show()
```

```python
greyimage=cv2.cvtColor(RGBimg, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(greyimage)
plt.axis(False)
plt.show()
```

```python
cv2.imshow('Greyscale image',greyimage)
cv2.waitKey(0)
```

```python
inverted_img=cv2.bitwise_not(greyimage)
cv2.imshow('Inverted Image',inverted_img)
cv2.waitKey(0)
```

```python
blur=cv2.GaussianBlur(inverted_img, (111,111),0)
```

```python
inverted_blur=cv2.bitwise_not(blur)
```

```python
cv2.imshow('Blur image',inverted_blur)
cv2.waitKey(0)
```

```python
imgSketch=cv2.divide(greyimage,inverted_blur, scale=256.0)
```

```python
cv2.imshow('Sketch',imgSketch)
cv2.waitKey(0)
```

```python
 cv2.imwrite('sketch.png',imgSketch)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.title('Actual image', size=20)
plt.imshow(RGBimg)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketched version', size=20)
RGBsketch=cv2.cvtColor(imgSketch, cv2.COLOR_BGR2RGB)
plt.imshow(RGBsketch)
plt.axis('off')
plt.show()
```
